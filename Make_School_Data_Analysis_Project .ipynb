```python

# will take all of the 2016 files ( yes i know theyre un sorted) and put them into one giant file, as a test for later interations
files = sorted(glob.iglob('2016/*.csv'))
header = OrderedDict()
data = []
for filename in files:
    with open(filename, 'r') as fin:
        csvin = csv.DictReader(fin)
        header.update(OrderedDict.fromkeys(csvin.fieldnames))
        data.append(next(csvin))
with open('output_filename_version2.csv', 'w', newline='') as fout:
    csvout = csv.DictWriter(fout, fieldnames=list(header))
    csvout.writeheader()
    csvout.writerows(data)
```

```python
#worked with several students to get this corrent and now submitting ( got the most help from jsutin since he's on my core project team,
#figuring out how to get the other cities, singapore might give me the most trouble)

# This was a hard dataset due tpo the fact that the data had to be scraped and cleaned and was not all the same size, 
#nor did it have the same columns, Had help from Kaichi, Justin and Tassos

location = 'datasets/SA Feedback Surveys_FINAL/2016/'

#Get all files in the 2016 folder
filenames = glob.glob(location + 'Anon Week * Feedback - *.csv')

#NOTE: SG IS FOR SINGAPORE-- was confused by that untill i started looking up abbreviations


#week 8 was previously having issues being read and had to be read it its own list


#settign for week 8
eigth_week = glob.glob(location + '*Week 8*')
#reads and addews to the csv files already established, so ti will look like the others
week8_df = pd.read_csv(eigth_week[0])


#List of names of columns scraped from the file names
df_2016 = pd.DataFrame([])
list_2016 = []

#print("FILE COUNT:", len(filenames))

#create week set and city set so there are no duplicate entries, important for the data frame
week_set = set()
city_set = set()

for file_name in filenames:
    rows = list()
    #split on spaces in file name
    file_data = file_name.split(" ")
    week = file_data[2]
    #split on . in file name
    city_list = file_data[-1].split('.')
    city = city_list[0]
    file_df = pd.read_csv(file_name)
    
    week_set.add(week)
    city_set.add(city)
    
    week_data = []
    city_data = []
    #addes the city and week data to the columns rows
    for _ in range(0, len(file_df)):
        week_data.append(week)
        city_data.append(city)
    #finally at the end the file data frames week and city property will have the week and city row data collected and added
    file_df["Week"] = week_data
    file_df["City"] = city_data
    list_2016.append(file_df)
    
clean_2016_df = pd.concat(list_2016)
# testing if it displays correctly
clean_2016_df

```
